```python
# Dealing with duplicate values 

# data_file

print(data_file.duplicated())
# Check if any dulpicate row is there or not 

print(data_file["country"].duplicated())              # { "country" - Primary Key }
# Check if the perticular column contains duplicate or not result is given row-wise
print(data_file["country"].duplicated().sum())

print(data_file.drop_duplicates())
# Removing all the duplicate Rows [Rows which are completely same] 

print(data_file.drop_duplicates("country"))
# Removing all the rows which have same column data
```

```python
# Working with missing data ( NULL Values - NaN - Not a Number - Empty Value )

print(data_file.isnull().sum())

# For Deleting all the null value 
print(data_file.dropna())

# Replacing NaN
import numpy as np
print(data_file.replace(np.nan, "1"))

data_file["beer_servings"] = data_file["beer_servings"].replace(np.nan, 0)


# Calculating Mean
print(data_file["spirit_servings"].mean())
data_file["spirit_servings"] = data_file["spirit_servings"].replace(np.nan, data_file["spirit_servings"].mean())
    # In this way your mean remains constant


# Forward Fill and Backward Fill
print(data_file.fillna(method = 'bfill'))        # Lower Value
print(data_file.fillna(method = 'ffill'))        # Upper Value
print(data_file.fillna('1'))
```
